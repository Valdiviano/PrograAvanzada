### **¿Qué es un paradigma de programación?** ###
Un *paradigma de programación* es un conjunto de principios, conceptos y métodos que establecen una forma específica de concebir y organizar el desarrollo de software. Los paradigmas de programación proporcionan un marco teórico y metodológico para abordar la solución de problemas computacionales, definiendo la manera en que los programadores deben estructurar y ejecutar sus programas.

Existen diversos paradigmas de programación, entre los cuales destacan los siguientes:

*   ***Programación imperativa:*** Se fundamenta en la especificación detallada de los pasos que el computador debe seguir para alcanzar un objetivo determinado, mediante el uso de instrucciones secuenciales, estructuras de control de flujo (como bucles y condicionales) y manipulación explícita del estado de las variables. Ejemplos de lenguajes imperativos incluyen C, Python (en su modalidad imperativa) y Java.
*   ***Programación funcional:*** Se basa en el uso de funciones matemáticas puras y la evitación de estados mutables y efectos colaterales, promoviendo un estilo de programación declarativo donde se enfatiza el "qué" se desea lograr en lugar del "cómo". Ejemplos de lenguajes funcionales son Haskell, Lisp, y Python (en su modalidad funcional).
*   ***Programación orientada a objetos (OOP)***: Organiza el código en unidades denominadas "objetos", que encapsulan tanto datos como métodos que operan sobre dichos datos, promoviendo la modularidad, la reutilización de código, y la representación natural de entidades del mundo real. Ejemplos de lenguajes orientados a objetos incluyen Java, C++, y Python (en su modalidad orientada a objetos).
*   ***Programación declarativa:*** Se enfoca en expresar el resultado deseado sin especificar detalladamente los pasos necesarios para alcanzarlo, delegando al lenguaje o al entorno de ejecución la responsabilidad de determinar cómo lograr dicho resultado. Ejemplos de lenguajes declarativos son SQL y Prolog.

### **¿En qué se basa la programación orientada a objetos?** ###
La *programación orientada a objetos* (POO) es un paradigma de programación que se fundamenta en la noción de "objetos", los cuales son abstracciones que representan entidades del mundo real o conceptual dentro de un sistema de software. Estos objetos se definen a través de "clases", que actúan como plantillas o moldes, encapsulando tanto los datos (atributos) como los procedimientos o funciones (métodos) que operan sobre dichos datos. La POO se basa en cuatro principios esenciales:

*   **Encapsulamiento:** Este principio establece que los datos y los métodos que los manipulan se agrupan en una entidad única (objeto), de tal manera que se ocultan los detalles internos de su implementación, exponiendo únicamente una interfaz pública. El encapsulamiento protege el estado interno del objeto contra modificaciones indebidas desde el exterior, lo cual garantiza la integridad de los datos y permite modificar la implementación interna sin afectar el código que depende de la interfaz del objeto.
*   ***Herencia:*** La herencia es un mecanismo que permite la creación de nuevas clases a partir de clases existentes, permitiendo a las subclases (clases derivadas) heredar atributos y métodos de las superclases (clases base). Este principio facilita la reutilización de código, la extensión de funcionalidades existentes y la organización jerárquica del software, promoviendo un diseño que refleja de manera más fiel las relaciones conceptuales entre los objetos.
*   ***Abstracción:*** La abstracción consiste en simplificar la complejidad de un sistema al reducirlo a sus características esenciales, omitiendo detalles irrelevantes desde el punto de vista del usuario o del desarrollador. A través de la abstracción, la POO facilita el manejo de la complejidad al permitir que los programadores se enfoquen en los aspectos significativos del problema a resolver, representando únicamente las propiedades y comportamientos esenciales de los objetos dentro del contexto del software.

La POO es utilizada ampliamente para desarrollar sistemas de software modulares, reutilizables y mantenibles, permitiendo modelar estructuras complejas de manera intuitiva y alineada con la realidad del dominio del problema. Este paradigma es particularmente adecuado para aplicaciones que requieren un alto grado de modularidad, extensibilidad y encapsulación.


### **¿Cuál es la diferencia entre recursividad e iteración, y cómo se relaciona esto con la notación Big O?** ###
La *recursividad* y la *iteración* son dos enfoques fundamentales en la programación para resolver problemas que requieren repetición de operaciones, pero difieren en su implementación y en su comportamiento en términos de eficiencia computacional.

*   ***Recursividad:*** La recursividad es una técnica en la cual una función se llama a sí misma para resolver subproblemas más pequeños de una versión mayor del problema original. La recursividad se basa en la definición de un caso base, que termina el proceso recursivo, y uno o más casos recursivos, que descomponen el problema en instancias más simples de sí mismo. Esta técnica es especialmente útil para problemas que tienen una estructura inherentemente jerárquica o que se pueden dividir de manera natural en subproblemas similares, como los algoritmos de búsqueda en árboles o la resolución de problemas de tipo "divide y vencerás".
*   ***Iteración:*** La iteración, en cambio, utiliza estructuras de control de flujo como bucles (*for* o *while*) para repetir una secuencia de instrucciones hasta que se cumpla una condición específica. A diferencia de la recursividad, la iteración no implica llamadas repetidas a una función, sino que ejecuta las instrucciones en un ciclo, lo cual generalmente requiere menos memoria, ya que no depende de la pila de llamadas del sistema.

####**Relación con la notación Big O:**####

La notación Big O se utiliza para expresar la *complejidad temporal* y *espacial* de los algoritmos, indicando cómo se escala el tiempo de ejecución o el uso de memoria a medida que crece el tamaño de la entrada.

*   ***Recursividad:*** Los algoritmos recursivos pueden ser menos eficientes en términos de complejidad espacial debido a la sobrecarga de la pila de llamadas, especialmente si tienen un alto número de niveles de recursión. Por ejemplo, un algoritmo recursivo simple como el cálculo de los números de Fibonacci sin optimización tiene una complejidad temporal de $O(2^n)$ y puede agotar rápidamente la memoria debido a las múltiples llamadas anidadas. Sin embargo, algunos algoritmos recursivos, como la búsqueda binaria, tienen una complejidad temporal de $O(\log n)$, que es más eficiente en términos de tiempo, aunque aún puede implicar una sobrecarga espacial de $O(\log n)$ en la pila de llamadas.
*   ***Iteración:*** Los algoritmos iterativos suelen ser más eficientes en términos de uso de memoria, ya que no requieren almacenamiento adicional en la pila de llamadas. Por ejemplo, un algoritmo iterativo que recorre una lista de tamaño $n$ tiene una complejidad temporal de $O(n)$ y una complejidad espacial de $O(1)$, ya que solo necesita una cantidad constante de espacio para almacenar las variables del bucle.

La recursividad suele ser más intuitiva y conveniente para ciertos tipos de problemas, pero puede ser ineficiente en términos de memoria y tiempo si no se maneja adecuadamente. La iteración, por su parte, es generalmente más eficiente en términos de espacio, pero puede requerir una mayor cantidad de código para problemas complejos. La elección entre recursividad e iteración depende de la naturaleza del problema, las restricciones de memoria y tiempo, y la claridad del código.


### **Explicar la diferencia de rendimiento entre 𝑂(1) y 𝑂(𝑛)** ###

La notación Big O se utiliza para describir la complejidad temporal de un algoritmo, es decir, cómo crece el tiempo de ejecución en función del tamaño de la entrada.

*   *$O(1)$* (Tiempo constante)*:* Un algoritmo con complejidad temporal $O(1)$ se ejecuta en un tiempo constante, independientemente del tamaño de la entrada. Esto significa que el tiempo de ejecución del algoritmo no aumenta cuando se incrementa el tamaño de los datos de entrada. Por ejemplo, acceder a un elemento en una lista (o arreglo) por su índice es una operación de tiempo constante, ya que se realiza en un solo paso sin importar cuántos elementos haya en la lista.

Un algoritmo $O(1)$ es extremadamente eficiente, ya que siempre tarda la misma cantidad de tiempo en completarse, sin importar el tamaño de los datos que se están procesando.

*   *$O(n)$* (Tiempo lineal)*:* Un algoritmo con complejidad temporal $O(n)$ tiene un tiempo de ejecución que crece de manera lineal con respecto al tamaño de la entrada. Esto significa que si la entrada se duplica, el tiempo de ejecución del algoritmo también se duplicará aproximadamente. Un ejemplo típico de un algoritmo con complejidad $O(n)$ es la búsqueda lineal en una lista no ordenada, donde se puede necesitar examinar cada elemento una vez para encontrar el elemento buscado o confirmar que no está presente.

En términos de rendimiento, un algoritmo $O(n)$ es menos eficiente que un algoritmo $O(1)$, especialmente a medida que el tamaño de la entrada aumenta. Sin embargo, sigue siendo manejable para tamaños de entrada moderados o cuando no es posible obtener un tiempo constante.


####**Diferencia de rendimiento:**

La diferencia de rendimiento entre \( O(1) \) y \( O(n) \) es significativa:


*   Un algoritmo \( O(1) \) es más rápido y eficiente para entradas grandes porque su tiempo de ejecución no depende del tamaño de la entrada. En el mejor de los casos, es posible realizar la operación en un tiempo fijo, independientemente del número de elementos a procesar.
*   Un algoritmo \( O(n) \), por otro lado, se vuelve más lento a medida que aumenta el tamaño de la entrada, ya que su tiempo de ejecución aumenta linealmente con el número de elementos. Este tipo de algoritmo puede ser ineficiente para entradas grandes, ya que requiere tiempo proporcional al tamaño de la entrada para completarse.

En resumen, mientras que $O(1)$ representa el escenario más eficiente en términos de tiempo de ejecución, \( O(n) \) representa un rendimiento lineal donde el tiempo de ejecución crece proporcionalmente con el tamaño de la entrada. Por lo tanto, siempre se preferirá un algoritmo de complejidad \( O(1) \) cuando sea posible, ya que ofrece un rendimiento significativamente mejor, especialmente para tamaños de entrada grandes.


### **¿Cómo se calcula el orden en un programa que funciona por etapas** ###
Para calcular la complejidad temporal de un programa que se ejecuta por etapas, es necesario analizar la complejidad de cada una de las etapas o componentes del programa y combinar dichas complejidades para obtener una expresión que refleje el comportamiento global del tiempo de ejecución del programa.

**Pasos para calcular la complejidad temporal de un programa por etapas:**



1.   ***Identificar y analizar cada etapa:*** Descomponer el programa en sus etapas o componentes lógicos, como funciones, bucles, llamadas a subrutinas, o cualquier otra operación relevante. Para cada etapa, se debe determinar su complejidad temporal en función del tamaño de la entrada, utilizando la notación Big O. Por ejemplo, si una etapa del programa implica recorrer un arreglo de tamaño $n$, la complejidad de esta etapa sería $O(n)$.

2.   ***Calcular la complejidad de cada etapa:*** Para cada etapa identificada, se determina su complejidad temporal específica. Esto puede incluir operaciones de tiempo constante $O(1)$, operaciones lineales $O(n)$, operaciones cuadráticas $O(n^2)$ entre otras. Es importante tener en cuenta el tipo de estructura de control utilizada (como bucles anidados o recursividad) al calcular la complejidad de cada etapa.


3.   ***Combinar las complejidades de las etapas:*** Una vez determinadas las complejidades temporales de todas las etapas, se deben combinar para obtener la complejidad global del programa. Existen dos casos principales:


*   ***Etapas secuenciales:*** Si las etapas del programa se ejecutan de forma secuencial (una después de otra), la complejidad temporal total es la suma de las complejidades de todas las etapas. Matemáticamente, esto se expresa como:

\begin{align*}
O(T) = O(T_1) + O(T_2) + \ldots + O(T_k)
\end{align*}

donde $T_i$ representa la complejidad temporal de la etapa $i$. En notación Big O, solo se considera el término de mayor magnitud, ya que domina el comportamiento del tiempo de ejecución para entradas grandes. Por ejemplo, si $O(T_1) = O(n)$ y $O(T_2) = O(n^2)$, entonces la complejidad total del programa es $O(n^2)$.

*   ***Etapas anidadas o dependientes:*** Si algunas etapas están anidadas (por ejemplo, bucles anidados) o son dependientes entre sí, la complejidad temporal global se calcula multiplicando las complejidades de estas etapas. Por ejemplo, si un bucle exterior de complejidad $O(n)$ contiene un bucle interior de complejidad $O(n)$, la complejidad combinada es $O(n) \cdot O(n) = O(n^2)$.


4.   ***Determinar la complejidad dominante:*** En la notación Big O, solo se incluye el término de mayor orden de crecimiento, ya que este término domina el tiempo de ejecución del programa a medida que el tamaño de la entrada crece. Por lo tanto, la complejidad global del programa se expresa en función de este término dominante.

Para calcular el orden de complejidad de un programa que funciona por etapas, se analiza cada etapa por separado, se determinan sus complejidades individuales y luego se combinan, considerando el tipo de ejecución (secuencial o anidada). Finalmente, se identifica la complejidad dominante para expresar el rendimiento global del programa en términos de notación *Big O*.

### ***¿Cómo se puede determinar la complejidad temporal de un algoritmo recursivo?***###

La *complejidad temporal* de un algoritmo recursivo describe cómo crece el tiempo de ejecución del algoritmo en función del tamaño de la entrada. Para determinar esta complejidad, se utiliza comúnmente una *relación de recurrencia*, que representa el tiempo de ejecución del algoritmo en términos del tiempo de ejecución de sus subproblemas más pequeños. A continuación se presentan los pasos generales para determinar la complejidad temporal de un algoritmo recursivo:


1.   ***Identificar el caso base y el caso recursivo:*** Un algoritmo recursivo se compone de uno o más casos base, que terminan la recursión, y uno o más casos recursivos, que dividen el problema original en subproblemas más pequeños. El caso base generalmente tiene una complejidad temporal constante $O(1)$.

2.   ***Formular la relación de recurrencia:*** La relación de recurrencia expresa el tiempo de ejecución $T(n)$ del algoritmo en función de:
*   El tiempo requerido para resolver los subproblemas recursivos.
*   El número de subproblemas en los que se divide el problema original.
*   El tiempo adicional requerido para combinar los resultados de los subproblemas o realizar cualquier operación fuera de las llamadas recursivas.

3.   ***Resolver la relación de recurrencia:*** Existen varios métodos para resolver la relación de recurrencia y determinar la complejidad temporal, como el *Método de Árbol Recursivo*, *Método de Sustitución o Expansión* y *Teorema Maestro*.

4.   ***Seleccionar el método adecuado:*** La elección del método para resolver la relación de recurrencia depende de la forma específica de la relación. El teorema maestro es una herramienta poderosa para relaciones de recurrencia que encajan en su formato, mientras que los métodos del árbol de recursión o de sustitución son más flexibles para casos más generales.

La complejidad temporal de un algoritmo recursivo se determina formulando una relación de recurrencia que refleje cómo el algoritmo divide el problema en subproblemas y resuelve cada uno de ellos, y luego resolviendo esta relación mediante métodos analíticos o el uso de teoremas específicos.

## **2) Programación con decoradores** ##

### **Descripción de lo que se va a realizar**

Se programa un código que meda y compare el tiempo de ejecución de dos métodos diferentes para calcular la cantidad de caminos posibles entre dos puntos en una PCB para diferentes tamaños. Utilizaremos un enfoque recursivo y un enfoque iterativo para resolver el problema. Los tiempos de ejecución de cada método se grafican para distintos tamaños de PCB, mostrando cómo varía la eficiencia entre ambos enfoques a medida que aumenta el tamaño de la grilla.

```python
import time
import matplotlib.pyplot as plt

# Decorador para medir y almacenar el tiempo de ejecución
def medir_tiempo(func):
    def wrapper(self, *args, **kwargs):
        start_time = time.time()  # Captura el tiempo de inicio
        resultado = func(self, *args, **kwargs)  # Ejecuta la función decorada
        end_time = time.time()  # Captura el tiempo de finalización
        # Almacena el tiempo de ejecución en un atributo de la instancia
        self.tiempo_ejecucion = end_time - start_time
        return resultado
    return wrapper

class PCBPathFinder:
    def __init__(self, N, M):
        """
        Inicializa la clase con el tamaño de la PCB (N x M).
        """
        self.N = N  # Número de filas
        self.M = M  # Número de columnas
        self.tiempo_ejecucion = 0  # Inicializa el atributo para almacenar el tiempo de ejecución

    def count_paths_recursive(self, x=0, y=0):
        """
        Método 1: Calcula la cantidad de caminos posibles usando un enfoque recursivo.
        """
        # Caso base: Si llegamos a la esquina inferior derecha (B)
        if x == self.N - 1 and y == self.M - 1:
            return 1

        # Caso base: Si se sale de los límites de la grilla
        if x >= self.N or y >= self.M:
            return 0

        # Recursión: Explorar hacia la derecha y hacia abajo
        caminos_derecha = self.count_paths_recursive(x + 1, y)
        caminos_abajo = self.count_paths_recursive(x, y + 1)

        return caminos_derecha + caminos_abajo

    def count_paths_iterative(self):
        """
        Método 2: Calcula la cantidad de caminos posibles usando un enfoque iterativo.
        """
        # Crear una tabla 2D para almacenar los resultados
        dp = [[0 for _ in range(self.M)] for _ in range(self.N)]

        # Inicializar la primera fila y la primera columna
        for i in range(self.N):
            dp[i][0] = 1
        for j in range(self.M):
            dp[0][j] = 1

        # Llenar la tabla dinámica
        for i in range(1, self.N):
            for j in range(1, self.M):
                dp[i][j] = dp[i - 1][j] + dp[i][j - 1]

        return dp[self.N - 1][self.M - 1]

    @medir_tiempo
    def compute_paths(self, method="iterative"):
        """
        Método que permite calcular la cantidad de caminos utilizando el método especificado.
        """
        if method == "recursive":
            return self.count_paths_recursive()
        elif method == "iterative":
            return self.count_paths_iterative()
        else:
            raise ValueError("Método desconocido. Use 'recursive' o 'iterative'.")

def generar_grafico_tiempos():
    # Considerar un rango más amplio de tamaños de grillas
    tamanos = list(range(5, 31, 2))  # PCB de tamaño 5x5 hasta 30x30, en pasos de 2
    tiempos_iterativo = []
    tiempos_recursivo = []

    for tamano in tamanos:
        pcb = PCBPathFinder(tamano, tamano)

        # Medir tiempo iterativo
        pcb.compute_paths(method="iterative")
        tiempos_iterativo.append(pcb.tiempo_ejecucion)

        # Medir tiempo recursivo solo para tamaños pequeños para evitar tiempos largos
        if tamano <= 15:  # Ajuste para evitar que la recursión se vuelva demasiado lenta
            pcb.compute_paths(method="recursive")
            tiempos_recursivo.append(pcb.tiempo_ejecucion)
        else:
            tiempos_recursivo.append(None)  # No calcular para tamaños grandes

    # Graficar los resultados
    plt.figure(figsize=(10, 6))
    plt.plot(tamanos, tiempos_iterativo, marker='o', label='Método Iterativo')

   # Excluir los puntos donde no se calcularon tiempos recursivos
    tamanos_recursivo = [t for t, tr in zip(tamanos, tiempos_recursivo) if tr is not None]
    tiempos_recursivo = [tr for tr in tiempos_recursivo if tr is not None]
    plt.plot(tamanos_recursivo, tiempos_recursivo, marker='o', label='Método Recursivo')

    plt.title('Tiempos de Ejecución para Métodos de Cálculo de Caminos')
    plt.xlabel('Tamaño de la PCB (N x N)')
    plt.ylabel('Tiempo de Ejecución (segundos)')
    plt.legend()
    plt.grid(True)
    plt.savefig('tiempos_ejecucion.svg')  # Guardar como archivo SVG
    plt.show()

# Generar y mostrar el gráfico
generar_grafico_tiempos()
```

```python
import time
import matplotlib.pyplot as plt

def medir_tiempo(func):
    """
    Decorador para medir el tiempo de ejecución de una función.
    """
    def wrapper(self, *args, **kwargs):
        start_time = time.time()
        resultado = func(self, *args, **kwargs)
        end_time = time.time()
        # Almacena el tiempo de ejecución en un atributo de la instancia
        tiempo = end_time - start_time
        print(f"Tiempo de ejecución de {func.__name__}: {tiempo:.6f} segundos")
        return resultado, tiempo
    return wrapper

class CaminoPCB:
    def __init__(self, filas, columnas):
        self.filas = filas
        self.columnas = columnas

    @medir_tiempo
    def calcular_caminos_iterativo(self):
        # Crear una matriz para almacenar los caminos posibles
        tabla = [[0] * self.columnas for _ in range(self.filas)]

        # Inicializar la primera fila y la primera columna
        for i in range(self.filas):
            tabla[i][0] = 1
        for j in range(self.columnas):
            tabla[0][j] = 1

        # Llenar la matriz calculando caminos acumulados
        for i in range(1, self.filas):
            for j in range(1, self.columnas):
                tabla[i][j] = tabla[i - 1][j] + tabla[i][j - 1]

        # Retornar el número de caminos hacia la esquina inferior derecha
        return tabla[self.filas - 1][self.columnas - 1]

    @medir_tiempo
    def calcular_caminos_recursivo(self, fila_actual=0, columna_actual=0):
        # Caso base: Si se alcanza la esquina inferior derecha
        if fila_actual == self.filas - 1 and columna_actual == self.columnas - 1:
            return 1

        # Inicializar contador de caminos
        total_caminos = 0

        # Verificar movimiento hacia abajo
        if fila_actual < self.filas - 1:
            total_caminos += self.calcular_caminos_recursivo(fila_actual + 1, columna_actual)[0]

        # Verificar movimiento hacia la derecha
        if columna_actual < self.columnas - 1:
            total_caminos += self.calcular_caminos_recursivo(fila_actual, columna_actual + 1)[0]

        return total_caminos

def graficar_tiempos():
    tamanos = list(range(2, 12))  # Tamaños de la PCB de 2x2 hasta 11x11
    tiempos_iterativo = []
    tiempos_recursivo = []

    for tamano in tamanos:
        pcb = CaminoPCB(tamano, tamano)

        # Medir tiempo iterativo
        _, tiempo_iterativo = pcb.calcular_caminos_iterativo()
        tiempos_iterativo.append(tiempo_iterativo)

        # Medir tiempo recursivo solo para tamaños pequeños para evitar tiempos largos
        if tamano <= 10:  # Ajuste para evitar que la recursión se vuelva demasiado lenta
            _, tiempo_recursivo = pcb.calcular_caminos_recursivo()
            tiempos_recursivo.append(tiempo_recursivo)
        else:
            tiempos_recursivo.append(None)  # No calcular para tamaños grandes

    # Graficar los resultados
    plt.figure(figsize=(10, 6))
    plt.plot(tamanos, tiempos_iterativo, marker='o', label='Método Iterativo')

    # Excluir los puntos donde no se calcularon tiempos recursivos
    tamanos_recursivo = [t for t, tr in zip(tamanos, tiempos_recursivo) if tr is not None]
    tiempos_recursivo = [tr for tr in tiempos_recursivo if tr is not None]
    plt.plot(tamanos_recursivo, tiempos_recursivo, marker='o', label='Método Recursivo')

    plt.title('Tiempos de Ejecución para Métodos de Cálculo de Caminos')
    plt.xlabel('Tamaño de la PCB (N x N)')
    plt.ylabel('Tiempo de Ejecución (segundos)')
    plt.legend()
    plt.grid(True)
    plt.savefig('tiempos_ejecucion.svg')  # Guardar como archivo SVG
    plt.show()

# Generar y mostrar el gráfico
graficar_tiempos()

```
