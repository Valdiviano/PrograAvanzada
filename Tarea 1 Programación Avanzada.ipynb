### **Descripci贸n de lo que se va a realizar**

Se programa un c贸digo que meda y compare el tiempo de ejecuci贸n de dos m茅todos diferentes para calcular la cantidad de caminos posibles entre dos puntos en una PCB para diferentes tama帽os. Utilizaremos un enfoque recursivo y un enfoque iterativo para resolver el problema. Los tiempos de ejecuci贸n de cada m茅todo se grafican para distintos tama帽os de PCB, mostrando c贸mo var铆a la eficiencia entre ambos enfoques a medida que aumenta el tama帽o de la grilla.

```python
import time
import matplotlib.pyplot as plt

def medir_tiempo(func):
    """
    Decorador para medir el tiempo de ejecuci贸n de una funci贸n.
    """
    def wrapper(self, *args, **kwargs):
        start_time = time.time()
        resultado = func(self, *args, **kwargs)
        end_time = time.time()
        # Almacena el tiempo de ejecuci贸n en un atributo de la instancia
        tiempo = end_time - start_time
        return resultado, tiempo
    return wrapper

class CaminoPCB:
    def __init__(self, filas, columnas):
        self.filas = filas
        self.columnas = columnas

    @medir_tiempo
    def calcular_caminos_iterativo(self):
        # Crear una matriz para almacenar los caminos posibles
        tabla = [[0] * self.columnas for _ in range(self.filas)]

        # Inicializar la primera fila y la primera columna
        for i in range(self.filas):
            tabla[i][0] = 1
        for j in range(self.columnas):
            tabla[0][j] = 1

        # Llenar la matriz calculando caminos acumulados
        for i in range(1, self.filas):
            for j in range(1, self.columnas):
                tabla[i][j] = tabla[i - 1][j] + tabla[i][j - 1]

        # Retornar el n煤mero de caminos hacia la esquina inferior derecha
        return tabla[self.filas - 1][self.columnas - 1]

    @medir_tiempo
    def calcular_caminos_recursivo(self, fila_actual=0, columna_actual=0):
        # Caso base: Si se alcanza la esquina inferior derecha
        if fila_actual == self.filas - 1 and columna_actual == self.columnas - 1:
            return 1

        # Inicializar contador de caminos
        total_caminos = 0

        # Verificar movimiento hacia abajo
        if fila_actual < self.filas - 1:
            total_caminos += self.calcular_caminos_recursivo(fila_actual + 1, columna_actual)[0]

        # Verificar movimiento hacia la derecha
        if columna_actual < self.columnas - 1:
            total_caminos += self.calcular_caminos_recursivo(fila_actual, columna_actual + 1)[0]

        return total_caminos

def graficar_tiempos():
    tamanos = list(range(2, 12))  # Tama帽os de la PCB de 2x2 hasta 11x11
    tiempos_iterativo = []
    tiempos_recursivo = []

    for tamano in tamanos:
        pcb = CaminoPCB(tamano, tamano)

        # Medir tiempo iterativo
        _, tiempo_iterativo = pcb.calcular_caminos_iterativo()
        tiempos_iterativo.append(tiempo_iterativo)

        # Medir tiempo recursivo solo para tama帽os peque帽os para evitar tiempos largos
        if tamano <= 10:  # Ajuste para evitar que la recursi贸n se vuelva demasiado lenta
            _, tiempo_recursivo = pcb.calcular_caminos_recursivo()
            tiempos_recursivo.append(tiempo_recursivo)
        else:
            tiempos_recursivo.append(None)  # No calcular para tama帽os grandes

    # Graficar los resultados
    plt.figure(figsize=(10, 6))
    plt.plot(tamanos, tiempos_iterativo, marker='o', label='M茅todo Iterativo')

    # Excluir los puntos donde no se calcularon tiempos recursivos
    tamanos_recursivo = [t for t, tr in zip(tamanos, tiempos_recursivo) if tr is not None]
    tiempos_recursivo = [tr for tr in tiempos_recursivo if tr is not None]
    plt.plot(tamanos_recursivo, tiempos_recursivo, marker='o', label='M茅todo Recursivo')

    plt.title('Tiempos de Ejecuci贸n para M茅todos de C谩lculo de Caminos')
    plt.xlabel('Tama帽o de la PCB (N x N)')
    plt.ylabel('Tiempo de Ejecuci贸n (segundos)')
    plt.legend()
    plt.grid(True)
    plt.savefig('tiempos_ejecucion.svg')  # Guardar como archivo SVG
    plt.show()

# Generar y mostrar el gr谩fico
graficar_tiempos()
```
